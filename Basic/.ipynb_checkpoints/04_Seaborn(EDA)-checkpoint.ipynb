```python
# 데이터 셋 - 양평구청 코로나 현황 (2020 ~ 2023)

covid=pd.read_csv(r'D:\Edu\Machine Learning\workspace\PythonBasic\Data\pandas\04_Yangcheon_COVID.csv', encoding='cp949')
covid.shape     # (1276, 2)
covid.head(5)   # 2020-03 ~ 2023-08
```

```python
covid.tail(5)
```

```python
covid.info()
```

```python
covid_copy=covid.copy()
```

```python
covid_copy['년월일']=pd.to_datetime(covid_copy['일자'])
covid_copy.info()
```

```python
covid_copy['년도']=covid_copy['년월일'].dt.year
covid_copy['월']=covid_copy['년월일'].dt.month
covid_copy['일']=covid_copy['년월일'].dt.day
covid_copy['요일']=covid_copy['년월일'].dt.day_name()
covid_copy.head()
```

```python
english_to_korean={                
    'Monday': '월요일',
    'Tuesday': '화요일',
    'Wednesday': '수요일',
    'Thursday': '목요일',
    'Friday': '금요일',
    'Saturday': '토요일',
    'Sunday': '일요일'
}
   
covid_copy['요일']=covid_copy['요일'].map(english_to_korean)
covid_copy.head()
```

```python
# 월별 확진자수
plt.figure(figsize=(10, 6))
sns.lineplot(data=covid_copy, x='월', y='확진자수', estimator=sum, errorbar=None)  # errorbar=오차
```

```python
# 년도별 확진자 수
plt.figure(figsize=(10, 5))
sns.lineplot(data=covid_copy, x='년월일', y='확진자수', estimator=sum, errorbar=None)
plt.xlabel("년도별")
```

```python
# 일별 확진자수
plt.figure(figsize=(10, 5))
sns.lineplot(data=covid_copy, x='일', y='확진자수', estimator=sum, errorbar=None)
```

```python
# 요일별 확진자수 
plt.figure(figsize=(10, 5))
sns.lineplot(data=covid_copy, x='요일', y='확진자수', estimator=sum, errorbar=None)
```

### 날짜 형식으로 변환

```python
copy_weather=weather.copy()
copy_weather.shape
```

```python
copy_weather['날짜형식']=pd.to_datetime(copy_weather['일시'])
copy_weather.info()

copy_weather['년도']=copy_weather['날짜형식'].dt.year
copy_weather['월']=copy_weather['날짜형식'].dt.month
copy_weather['일']=copy_weather['날짜형식'].dt.day
copy_weather.head()
```

```python
copy_weather.columns
```

### 시각화
* 숫자형(시간형) / 숫자형 

```python
# 마이스너 설정
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10, 5))
plt.title('서울시 2023년도 기온 변화')

sns.lineplot(data=copy_weather, x='날짜형식', y='평균기온(°C)', errorbar=None, label='평균기온')
sns.lineplot(data=copy_weather, x='날짜형식', y='최저기온(°C)', errorbar=None, label='최저기온')
sns.lineplot(data=copy_weather, x='날짜형식', y='최고기온(°C)', errorbar=None, label='최고기온')

plt.xlabel('일별')
plt.ylabel('기온')

plt.show()
```

```python
spring=copy_weather.query('월 == 3')        # spring=copy_weather[copy_weather['월'] == 3]
spring.head()
spring.shape
```

```python
plt.figure(figsize=(10, 3))
plt.title('서울시 2023년도 3월의 기온 변화')

sns.lineplot(data=spring, x='날짜형식', y='평균기온(°C)', errorbar=None, label='평균기온')
sns.lineplot(data=spring, x='날짜형식', y='최저기온(°C)', errorbar=None, label='최저기온')
sns.lineplot(data=spring, x='날짜형식', y='최고기온(°C)', errorbar=None, label='최고기온')

plt.xlabel('일')
plt.ylabel('기온')

plt.show()
```
