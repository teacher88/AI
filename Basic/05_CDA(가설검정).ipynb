# CDA
- CDA 확증적 데이터 분석 / Confirmatory Data Analysis 
- 통계적 가설 검정 기법
- 통계절 가설 검증
  - 귀무가설(H0): 대립 가설의 반대 가설 / 보통 가설, 일반적인 상태
  - 대립가설(H1): 내가 의심하는 상태 / 내가 규명하고자 하는 가설
  - p-value (확률값)를 통해서 가설이 참/거짓 인지 확인하는 분석기법
    - 귀무 가설이 참일 확률 0 ~ 1
    - 유의 수준 : 가설 검정 판단의 기준 0.05
      - p-value > 0.05 : 귀무 가설 참 (귀무 가설 기각 실패)
      - p-value < 0.05 : 대립 가설 참 (귀무 가설 기각)
- 예) 가설 : 고객 성별에 따라 매출액에 차이가 있는가?
  - 귀무 가설: 고객 성별에 따른 매출 차이가 없다.
  - 대립 가설: 고객 성별에 따른 매출 차이가 있다.

  - 대림가설: 해당 연속형 데이터 분포는 정규 분포 아니다.

#### One Sample T-Test (일 표본 평균 검정)
- 하나의 값과 하나의 집단의 평균 검정
    - 귀무 가설 : A값은 집단의 평균과 같다.
    - 대림 가설 : A값은 집단의 평균과 다르다.
- Ex1) 수능 시험 점수 결과 좋다? 과연 그런가?
- Ex2) 집단 선택 : 학교
    - 기준 설정 : 학교 평균 점수 70점
    - 평균 계산 : 우리반 평균 계산 70점
    - 비교 : 우리반이 평균보다 잘했는가? 못했는가?
    - 통계 검정 : T-Test 평균의 차이가 있는지 없는지

```python
# 남녀 데이터프레임
print(tips['sex'].value_counts())

tips_m=tips[tips['sex']=='Male']
tips_f=tips[tips['sex']=='Female']
tips_m.shape, tips_f.shape
```

```python
print(tips['tip'].mean())
tips.groupby('sex', observed=False)['tip'].mean()
```

```python
# 하나의 값과 하나의 집단의 평균 검정
# 귀무가설: 팁 평균(2.9)과 남자손님 팁 평균 차이가 없다. 
# 대립가설: 팁 평균(2.9)과 남자손님 팁 평균 차이가 있다.

stats.ttest_1samp(tips_m['tip'], tips['tip'].mean())      # (표본데이터, 모집단 평균)
# pvalue 0.44 > 0.05 유의수준 / tip 차이가 없다 / 귀무가설 참, 귀무가설 기각 실패
```

```python
stats.ttest_1samp(tips_f['tip'], tips['tip'].mean())
# pvalue 0.18 > 0.05 유의수준 / tip 평균 차이가 없다 / 귀무가설 참, 귀무가설 기각 실패
```

### 문자형 - 비율 검정
- Z-Test 

- 일 표본 비율 검정 (One-Proportions Z-Test)
    - 특정 모집단의 비율이 기대값(기준값)과 다른지 검정
    - 귀무가설(H₀): 표본 비율 = 기준 비율의 차이가 없다

```python
busan=pd.read_csv('./Data/pandas/06_Busan.csv', encoding='cp949')
busan.shape
```

```python
busan.head()
```

```python
busan.info()
```

```python
busan['점수'].describe()
```

```python
jumsu=busan[busan['점수'] == 100]
jumsu.shape
```

```python
count=jumsu['점수'].count()              # 응답한 사람 수
nobs=busan['점수'].count()               # 전체 샘플 크기
prop=0.9                                # 기준 비율 (예: 50%)
count, nobs, prop
```

```python
#  귀무가설(H₀): 표본 비율=기준 비율의 차이가 없다.
#  귀무가설(H₀): 표본 비율=기준 비율의 차이가 있다.

proportions_ztest(count, nobs, prop)
# pvalue 0.640 > 0.05 / 비율의 차이가 없다. 즉 만족도 95% / 귀무가설 참, 귀무가설 기각 실패

# Ex) prop=0.95 1.6614027532948853e-31 
```

- 두 표본 비율 검정 (Two-Proportions Z-Test)
    - 남성 100명 중 60명이 특정 설문에 응답하고, 여성 120명 중 70명이 응답한 경우,
    - 비율에 차이가 있는지?
      - 귀무가설: 비율에 차이가 없다.
      - 대립가설: 비율에 차이가 있다.

```python
m_men=100        # 남성 100명 
answer_men=60    # 만족도(9점), 60명

m_women=120       # 여성 120명
answer_women=70   # 만족도(9점), 70명 

counts=[answer_men, answer_women]  # 응답 수
nobs=[m_men, m_women]              # 만족도 (9명)

# z-test 수행
proportions_ztest(counts, nobs)  
# pvalue 0.80 > 0.05 유의수준 / 설문 만족도 (비율)에 차이가 없다 / 귀무가설 참, 귀무가설 기각 실패
```

## 다변수 검정
### X(문자형) / Y(숫자형) 
* 두 집단의 평균 차이가 있는지 검정
  - 귀무가설: 두 집단 간 평균의 차이가 없다.
  - 대립가설: 두 집단 간 평균의 차이가 있다.

- X(문자형)/Y(문자형): Chi Squere Test / 성별에 따른 합격 여부 차이가 있는가?
- X(숫자형)/Y(숫자형): 키가 크면 몸무게가 증가한다. 연봉이 오르면 인사고과 점수가 올라갔는가? >> 예측
    - 상관분석: 정규 Person, 비정규 Spearman
    - 회귀분석 >> 회귀 예측 모델
- X(숫자형)/Y(문자형): 로지스틱분석 >> 분류 예측 모델

- 성별 여부에 따른 팁 차이 
    - 남녀 두 집단간 팁의 평균이 차이가 있는가?
    - 남녀 그룹별 팁 평균과 그래프로 먼저 확인

```python
# 절차 >> 정규/비정규 >> 정규(등분산, 이분산) >> 등분산(Two Sample T-Test, ANONA Test) : 여러 집단의 분산이 동일
#                                          >> 이분산(Two Sample T-Test, Kruskal Wills Test) : 여러 집단의 분산이 동일하지 않음
#                    >> 비정규(Wilcoxon Test, Kruskal Wills Test)

# 남녀 두 집단의 팁의 평균이 차이가 있는가?
# 귀무가설: 남녀간 팁 평균의 차이가 없다.
# 대립가설: 남녀간 팁 평균의 차이가 있다.

# 정규성 검정
stats.normaltest(tips['tip'])   # pvalue=5.796294322907102e-18 비정규 - Wilcoxon Test
```

```python
# 남녀 데이터 프레임
tips_m=tips[tips['sex']=='Male']            # tips.query("sex=='Male'")
tips_f=tips[tips['sex']=='Female']          # tips.query("sex=='Female'")

# 2집단(남, 녀) - Wilcoxon Test(윌콕슨 검정) / 3집단 Kruskal Wills Test(크루스칼 왈리스 검정)
stats.ranksums(tips_m['tip'], tips_f['tip'])

# 결과] pvalue 0.38 > 0.05 유의수준 / 귀무가설 참, 귀무가설 기각 실패, 통계적 근거 없음, 남녀간 팁 평균의 차이가 없다.
```

#### 정규
* 등분산 검정 - Leven Test
    - 귀무가설:남/여 그룹의 tip값에 분산의 차이가 없다. (분산 동일)
    - 대립가설:남/여 그룹의 tip값에 분산의 차이가 있다. (분산 동일 하지 않음)

```python
# 가정 : 정규성 
stats.normaltest(tips['tip'])  

# ***************** 현재 데이터가 없으니 정규성이라고 가정하고 실습해 보자 ******************
# 정규성 >> 등분산 >> 2집단 two sample t test /  ttest_ind(, equal_val=True), 3집단 이상 ANOVA test / f_oneway()
#          이분산 >> 2집단 two sample t test /  ttest_ind(, equal_val=False), 3집단 이상 Kruskal Wills Test / kruskal()
```

```python
tips_m=tips[tips['sex']=='Male']
tips_f=tips[tips['sex']=='Female']

# 등분산 검정 - 귀무가설(분산의 차이가 없다), 대립가설(분산의 차이가 있다)
stats.levene(tips_m['tip'], tips_f['tip'])

# 결론) pvalue 0.15 > 0.05 유의수준 / 귀무가설 참, 귀무가설 기각 실패 / 남여 그룹의 tip 분산의 차이가 없다. 통계적 근거 없음
```

```python
sns.histplot(data=tips_m, x='tip', kde=True)
sns.histplot(data=tips_f, x='tip', kde=True)

# tips_m['tip'].var(), tips_f['tip'].var()      # 2.2174242119875878, 1.3444275060144346
```

```python
# 등분산 동일 >> Two Sample T-Test
# 귀무가설: 남여 그룹의 tip 평균값의 차이가 없다.
# 대립가설: 남여 그룹의 tip 평균값의 차이가 있다.

stats.ttest_ind(tips_m['tip'], tips_f['tip'], equal_var=True)   # equal_var=True/False

# 결론) pvalue 0.16 > 0.05 유의수준 / 귀무가설 참, 귀무가설 기각 실패 / 두 집단의 평균의 차이가 없다. 통계적 근거 없음
```

###  X(문자형) / Y(문자형)
#### Chi Square Test
- 두 범주형 집단의 독립성 검정
- 귀무가설: A항목과 B항목은 서로 독립적이다. (연관성이 없다)
- 대립가설: A항목과 B항목은 서로 의존적이다. (연관성이 있다)

- 성별에 따른 흡연의 차이
- 그룹화, 시각화, 검정

```python
tips['sex'].value_counts()
```

```python
tips['smoker'].value_counts()
```

```python
tips.groupby(['sex','smoker'], observed=False)['smoker'].size()
```

```python
sns.countplot(data=tips, x='sex', hue='smoker')
```

```python
df_crosstab=pd.crosstab(tips['sex'], tips['smoker'])
df_crosstab        # 교차표
```

```python
# 귀무가설: 성별에 따라 흡연의 차이가 없다. (서로 독립)
# 대립가설: 성별에 따라 흡연의 차이가 있다. (서로 연관)

stats.chi2_contingency(df_crosstab)     

# 결론] pvalue 1.0 > 0.05 유의수준 / 귀무가설 참, 귀무가설 기각 실패, 성별에 따라 흡연의 차이 없다.
# 결과값] 통계량, pvalue, 자유도, 추정값(기대 빈도)
```
