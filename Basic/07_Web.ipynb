### 카카오 맵

```python
def get_kakao(query):
    url='https://dapi.kakao.com/v2/local/search/address.json'
    headers={'Authorization':'KakaoAK 94d9716fb80cc7808757dad0241b4370'}
    query={'query': query}
    kakao=requests.get(url, headers=headers, params=query).json()
    # print(kakao)
    
    address=kakao['documents'][0]['address']
    lng=address['x']    # 경도
    lat=address['y']    # 위도
    return lat, lng
    
get_kakao('종로구서린동70 알파빌딩')
```

```python
def get_latlng(address, sigungu):    
    url='https://dapi.kakao.com/v2/local/search/address.json'
    headers={'Authorization':'KakaoAK 94d9716fb80cc7808757dad0241b4370'}
    query={'query': address}
    kakao=requests.get(url, headers=headers, params=query).json()

    try:
        address=kakao['documents'][0]['address']
        lng=address['x']    # 경도
        lat=address['y']    # 위도
    except:
        lng=None 
        lat=None
    return lat, lng
```

```python
for idx, address in enumerate(corona_df['주소']):
     lat, lng=get_latlng(address, corona_df.loc[idx, '시군구'])
     corona_df.loc[idx, 'lat']=lat
     corona_df.loc[idx, 'lng']=lng
```
