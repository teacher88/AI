# Decision Tree
## DecisionTree Classifier
* 비선형 관계 학습
* 해석이 쉽고 직관적인 모델
* 스케일링이 필요없음
* 적은 전처리 필요 (결측치)
* 다양한 데이터 타입 지원 (수치형과 범주형 변수)

```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib as mpl  
mpl.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False  

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

### 와인 품종 분류- Logistic Regression
- N = 6497 * 4
- 설명 변수
    - alcohol : 알코올 도수 
    - sugar : 설탕 함류
    - pH : 산도(농도)
- 목표 변수
    - 레드와인 0, 화이트 와인 1

```python
wine=pd.read_csv('./Data/supervised/wine.csv')
wine.shape
```

```python
wine.head()

# data : 알코올 도수, 당도, PH(농도)
# target : 1(양성) 클래스는 화이트 와인
#          0(음성) 클래스는 레드 와인
```

```python
wine.info()      # 데이터 정보
```

```python
wine.describe().T    # 기술 통계량
```

```python
wine['class'].value_counts()    # 화이트, 레드 
```

```python
sns.pairplot(data=wine, hue='class')    # 시각화
```

```python
wine.corr().round(2)           # 상관 관계
# sns.heatmap(wine.corr(), annot=True)
```

```python
# 설명, 목표 변수 설정
X=wine.iloc[:, :3]
Y=wine.iloc[:, -1]
X.shape, Y.shape
```

```python
# 학습 및 평가
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, random_state=42)

scaler=StandardScaler()
scaler.fit(X_train)

X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)

lr=LogisticRegression()
lr.fit(X_train_scaled, Y_train)

print("학습:", lr.score(X_train_scaled, Y_train))
print("일반화:", lr.score(X_test_scaled, Y_test))
```

```python
# 평가 지표 : 정확도, 정밀도, 재현율, f1스코어
Y_test_pred=lr.predict(X_test_scaled)
print(classification_report(Y_test, Y_test_pred))  # 1(양성)  화이트 와인,  0(음성) 레드 와인
```

```python
print(lr.coef_, lr.intercept_)
```

```python
# 결과 
# 0.51268071  1.67335441  -0.70910354 
# 알코올도수  당도         PH(농도, 산도)

# 추측) 알코올 도수와 당도가 높을 수록 화이트 와인 가능성이 높다.
#       PH가 높을 수록 레드 와인 가능성이 높다 
#      기울기의 값과 절편값으로  이 숫자가 무엇을 의미인지 설명하긴 어렵다.
```

### 와인 품종 분류 - DecisionTreeClassifier
* 개념
  - 분류와 회귀가 모든 가능한 머신러닝 알고리즘
  - 질문을 던져서 맞고 틀리는 것에 따라 대상이 좁혀 나감
  - RandomForest 기본 구성 요소 <br><br>

* 구조
  - Root Node : 모든 데이터 정보를 갖고 있음
  - Decision Node : 데이터를 분류할 질문들이 들어감
  - lef Node : 의사 결정을 하지 않는 가장 마지막(노드 맨 끝)에 존재
  - depth : 깊이를 조정하며 과대적합을 예방 <br><br>

* X : 분기 기준
  - 테스트 조건은 평균값
  - 속성은 상관도 sugar > alcohol > ph  <br><br>

*  Y : 평가 기준
   - 수학적 계산 : 지니인덱스, 엔트로피(로그)
   - 불순도(gini)가 작아지는 방향 / ERROR
   - 데이터 분할 기준 (계산 방법 PPT) <br><br>
    
* 데이터 분할 할때 마다 불순도 작아짐 즉, 정보이득 

```python
dt=DecisionTreeClassifier(random_state=42)
dt.fit(X_train_scaled, Y_train)

print("학습:", dt.score(X_train_scaled, Y_train))
print("일반화:", dt.score(X_test_scaled, Y_test))

# 과대 적합
```

```python
plt.figure(figsize=(10, 7))    #  너비가 10인치, 높이가 7인치
plot_tree(dt)
plt.show()

# 리프 노드 : 다수(분류), 평균(회귀)
```

### 지니 불순도
* 노드에서 데이터를 분할할 기준 정함

```python
plt.figure(figsize=(10, 7))

# filled=True 지정하면 클래스마다 색깔를 부여한다.
plot_tree(dt, max_depth=1, filled=True, feature_names=['alochol', 'suger', 'pH']) 
plt.show()
```

```python
# 지니 불순도 계산
# 1-(음성클래스 비율 ^2 + 양성 클래스 비율 ^2)

print("부모 불순도:", 1-((1165/4872)**2 + (3707/4872)**2))
print("왼쪽 True:", 1-((1080/2690)**2 + (1610/2690)**2))
print("오른쪽 False", 1-((85/2182)**2 + (2097/2182)**2))
```

# 모델 최적화
* 교차검증, 하이퍼파라미터

```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib as mpl  
mpl.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

# 전처리 부분
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_validate

# 모델
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

# 하이퍼파라메타
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

## 교차검증

```python
# from sklearn.model_selection import cross_validate

wine=pd.read_csv('./Data/supervised/wine.csv')
wine.shape
```

```python
X=wine.iloc[:, :3]
Y=wine.iloc[:, -1]

# X.head()
# Y.head()
```

```python
# train & test dataset : default 70 : 20
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2, random_state=42)

# 표준화
scaler=StandardScaler()
scaler.fit(X_train)

X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)

# 모델 학습
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, Y_train)
```

```python
# 5-fold 교차 검증을 사용하여 모델 평가
validate_score=cross_validate(knn, X_train_scaled, Y_train, cv=5)   # default=5
validate_score
```

- fit_time : 모델 학습에 소요된 시간
- score_time :  테스트 데이터에 대한 점수를 계산하는 데 걸린 시간
- test_score : 데이터에 대해 모델의 성능을 평가한 정확도
- scoring='f1', 'precision', 'recall' 사용가능  / scoring = ['f1', 'precision', 'recall']

```python
print("학습:", knn.score(X_train_scaled, Y_train))
print("검증:", validate_score['test_score'].mean())
print("일반화:", knn.score(X_test_scaled, Y_test))
```

```python
Y_test_pred=knn.predict(X_test_scaled)
print(classification_report(Y_test, Y_test_pred))
```

## HyperParameterTuning
* 모델 파라미터 : 머신러닝 모델이 학습하는 파라미터
* 하이퍼파라미터 : 모델이 학습할 수 없어서 사용자가 지정하는 것
  - 클래스나 매개변수로 표현
  - 작업은 라이브러리가 제공하는 기본 값으로 사용
  - 그 다음 검증 세트의 점수나 고차 검증을 통해서 매개변수를 조금씩 바꿈
  - 모델 마다 매개 변수 적게 1~2개, 많게는 5~6개 

### GridSerach
* 하이퍼파라미터 탐색과 교차 검증을 한번에 수행 <br>
  (하이퍼파라미터 범위 설정, 탐색 실행, 최적 조합 선택)
* 사용 방법
  - 하이퍼파라미터 범위 설정
  - 탐색 실행 (교차 검증)
  - 최적 조합

```python
from sklearn.model_selection import GridSearchCV
```

#### KNN

```python
# 학습 & 테스트 
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2, random_state=42)

scaler=StandardScaler()
scaler.fit(X_train)

# 표준화
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)

# 모델 학습
# 1) 탐색할 파라미터 그리드 설정
params={'n_neighbors' : [3, 5, 7, 9, 11]}

# 2) GridSearchCV 초기화
#    5 * 5 = 25번 모델 훈련 / 병렬처리 CPU 코어 -1은 모든 코어 사용
gs=GridSearchCV(KNeighborsClassifier(), params, cv=5, n_jobs=-1)

# 3) GridSearchCV를 통해 최적의 파라미터 탐색
gs.fit(X_train_scaled, Y_train)

print("최적의 매개변수:", gs.best_params_)       # 최적의 하이퍼파라미터 조합을 출력
print("최적의 교차검증점수:", gs.best_score_)    # 그 최적의 조합에서 교차 검증으로 얻어진 평균 점수     

# 4) 테스트 데이터로 최적 모델 평가
knn_best=gs.best_estimator_                   # 최적의 하이퍼파라미터 설정으로 학습된 모델을 반환
knn_best.score(X_test_scaled, Y_test)
```

```python
Y_test_pred=gs.predict(X_test_scaled)
print(classification_report(Y_test, Y_test_pred))
```

#### DecisionTree
* criterion: 분할 기준을 지정. "gini" 또는 "entropy" / gini
* splitter: 노드를 분할하는 전략을 선택. "best" 또는 "random" / best
* max_depth: 트리의 최대 깊이를 제한 / none
* min_samples_split: 노드를 분할하기 위한 최소 샘플 수 / 2
* min_samples_leaf: 리프 노드에 있어야 할 최소 샘플 수 / 1
* max_features: 각 노드에서 분할에 사용할 최대 특성 수 / 모든 특성
* max_leaf_nodes: 리프 노드의 최대 개수를 지정 / none
* min_impurity_decrease: 분할을 수행하는 것이 불순도 감소량이 지정된 값보다 큰 경우 / 0보다 커야 함

```python
from sklearn.model_selection import GridSearchCV

# 학습 & 테스트 
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2, random_state=42)

# 모델 학습
# 1) 탐색할 파라미터 그리드 설정   임퓨어리티 디크리스
params={'max_depth': range(5, 20, 1),     # 트리의 최대 깊이
        'min_impurity_decrease': np.arange(0.0001, 0.001, 0.0001),   # 실수, 불순도가 감소할 때만 분할 가능 (예..최소 0.01만큼)                              
        'min_samples_split': range(10, 100, 10)}                      # 노드 분할에 필요한 최소 샘플수가 있어야 분할 가능

# 2) GridSearchCV 초기화
gs=GridSearchCV(DecisionTreeClassifier(random_state=42), params, n_jobs=-1)
gs.fit(X_train, Y_train)
```

```python
# 3) GridSearchCV를 통해 최적의 파라미터 탐색
print(gs.best_params_)             # 최적의 하이퍼파라미터 조합을 출력
print(gs.best_score_)              # 그 최적의 조합에서 교차 검증으로 얻어진 평균 점수  

# 4) 테스트 데이터로 최적 모델 평가
dt_best=gs.best_estimator_         # 최적의 하이퍼파라미터 설정으로 학습된 모델을 반환
dt_best.score(X_test, Y_test)      # 정확도

# 정리] 
# max_depth: 14, min_impurity_decrease : 9, min_samples_split :10
# 9 * 14 * 10 = 1260* 5(fold) = 6300번 모델 학습
```

```python
Y_test_pred=gs.predict(X_test)
print(classification_report(Y_test, Y_test_pred))
```

# Ensemble
* 앙상블(Ensemble)
    - 여러 개별 모델을 결합하여 하나의 강력한 예측 모델을 구축하는 기법
    - 다른 종류의 모델을 결합하여 앙상블을 구성 (결정트리, 로지스틱 회귀, KNN 등)
    - 주로 분류 문제에서 많이 사용되지만, 회귀 문제에서도 가능함
    - 결국,  Decision Tree Overfitting 방지 <br><br>

* 결합 방법
    - 보팅(Voting) : 서로 다른 알고리즘을 가진 여러 개의 모델을 사용하여 예측을 결합 <br>
                     Hard 보팅 (다수결), Soft 보팅 (예측 확률 평균)
    - 배팅(Bagging) : 동일한 알고리즘을 사용하지만 훈련 데이터셋을 무작위로 부분 샘플링하여 <br>
                여러 개의 부분 모델을 학습, 평균 또는 다수결을 통해 예측 (랜덤 포레스트)
    - 부스팅(Boosting) : 순차적으로 약한 학습기(weak learner)를 학습시켜 <br>
                        강력한 학습기(strong learner)를 만듬. <br>
                        예측 오류에 가중치를 부여하여 반복적으로 모델을 학습 (그래디언트 부스팅, 에이다부스트, XGBoost, CatBoosting)

## Voting
* 서로 다른 알고리즘을 가진 여러 개의 모델을 사용하여 예측
* Hard 보팅 (다수결), Soft 보팅 (예측 확률 평균)
* Soft 보팅 상대적으로 우수

```python
# 통계량
cancer_df.describe()
```

```python
cancer_df['label'].value_counts()
```

```python
corr_data=cancer_df.corr()['label'].abs()
corr_sort=corr_data.sort_values(ascending=False)
corr_sort    # 시리즈로 반환
```

```python
len(corr_sort.index)
```

```python
# 입력 데이터 수정
corr_df=corr_sort[1:26]
corr_df
```

```python
# 입력 & 정답 데이터
idx=corr_df.index
idx

X=cancer_df[idx]
X

Y=cancer_df['label']
Y

X.shape, Y.shape
```

```python
# 학습 & 테스트 셋
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, random_state=42)

# 표준화
scaler=StandardScaler()
scaler.fit(X_train)

X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

### 모델 학습

```python
# 개별 모델 정의
knn=KNeighborsClassifier()
lr=LogisticRegression()
dt=DecisionTreeClassifier()

# 보팅 분류기 정의 (분류기를 구분하기 위해 사용한 임의의 문자열, 객체)
vc=VotingClassifier(estimators=[('KNN', knn), ('LR', lr), ('DT', dt)], voting='soft')     # 확률 평균
# vc=VotingClassifier(estimators=[('KNN', knn), ('LR', lr), ('DT', dt)], voting='hard')   

vc.fit(X_train_scaled, Y_train)
vc.score(X_train_scaled, Y_train), vc.score(X_test_scaled, Y_test)
```

### 교차 검증

```python
# 5-fold 교차 검증을 사용하여 모델 평가
scores=cross_validate(vc, X_train_scaled, Y_train, return_train_score=True, cv=5)   # default=5

print("학습 5-Fold 교차 검증 점수:", scores['train_score'])
print("학습 교차 검증 평균:", np.mean(scores['train_score']))
print()

print("일반화 5-Fold 교차 검증 점수:", scores['test_score'])
print("일반화 교차 검증 평균:", np.mean(scores['test_score']))
```

```python
# 개별 모델들의 교차 검증 점수 계산 및 출력
clfs=[knn, lr, dt]
for clf in clfs:
    class_name=clf.__class__.__name__
    scores=cross_val_score(clf, X_train_scaled, Y_train, cv=5)
    
    print(class_name)
    print("5-Fold 교차 검증 점수:", scores)
    print("교차 검증 평균:", np.mean(scores), "\n")
```

```python
# DecisionTre
dt1=DecisionTreeClassifier(max_depth=5)
dt2=DecisionTreeClassifier(max_depth=5)
dt3=DecisionTreeClassifier(max_depth=5)

vc=VotingClassifier(estimators=[('dt1', dt1), ('dt2', dt2), ('dt3', dt3)], voting='hard')   # 다수결
vc.fit(X_train_scaled, Y_train)
vc.score(X_train_scaled, Y_train), vc.score(X_test_scaled, Y_test)
```
